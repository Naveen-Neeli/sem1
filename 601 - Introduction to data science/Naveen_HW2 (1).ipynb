```python
#2
import csv
import numpy as np
from matplotlib import pyplot as plt
f2 = open ("C:\\Users\\A D M I N\\Desktop\\601-Intro to Data Science\\Assignments\\Asnmt-2\\Messi.csv","r")
file2 = csv.DictReader(f2)
seasonstartm = []
seasonendm = []
clubappm = []
clubgoalm = []
natappm = []
natgoalm = []
for col in file2:
    seasonstartm.append(int(col['SeasonStart']))
    seasonendm.append(int(col['SeasonEnd']))
    clubappm.append(int(col['ClubApp']))
    clubgoalm.append(int(col['ClubGoals']))
    natappm.append(int(col['NatApp']))
    natgoalm.append(int(col['NatGoals']))

f3 = open ("C:\\Users\\A D M I N\\Desktop\\601-Intro to Data Science\\Assignments\\Asnmt-2\\Ronaldo.csv","r")
file3 = csv.DictReader(f3)
seasonstartr = []
seasonendr = []
clubappr = []
clubgoalr = []
natappr = []
natgoalr = []
for col in file3:
    seasonstartr.append(int(col['SeasonStart']))
    seasonendr.append(int(col['SeasonEnd']))
    clubappr.append(int(col['ClubApp']))
    clubgoalr.append(int(col['ClubGoals']))
    natappr.append(int(col['NatApp']))
    natgoalr.append(int(col['NatGoals']))
```

```python
plt.plot(seasonstartm,clubappm, label='club')
plt.plot(seasonstartm,natappm, label='national')
plt.title('Messi: Appearances')
plt.legend(frameon=True)

```

```python
plt.plot(seasonstartr,clubappr, label='club')
plt.plot(seasonstartr,natappr, label='national')
plt.title('Ronaldo: Appearances')
plt.legend(frameon=True)
```

```python
plt.plot(seasonstartm,clubgoalm, label='club')
plt.plot(seasonstartm,natgoalm, label='national')
plt.title('Messi: Goals')
plt.legend(frameon=True)
```

```python
plt.plot(seasonstartr,clubgoalr, label='club')
plt.plot(seasonstartr,natgoalr, label='national')
plt.title('Ronaldo: Goals')
plt.legend(frameon=True)
```

```python
messi_totalgames = 0
ronaldo_totalgames = 0
tgm = 0
tgr = 0
for i in range(0,18):
    tgm +=  clubgoalm[i]
    tgm += natgoalm[i]
    tgr += clubgoalr[i]
    tgr += natgoalr[i]
    messi_totalgames += clubappm[i]
    messi_totalgames += natappm[i]
    ronaldo_totalgames += clubappr[i]
    ronaldo_totalgames += natappr[i]
    
print("messi goals:", tgm)
print("ronaldo goals:", tgr)
print("messi total games:", messi_totalgames)
print("ronaldo total games:", ronaldo_totalgames)
messi_gpg = tgm / messi_totalgames
ronaldo_gpg = tgr / ronaldo_totalgames
print("messi goals:", messi_gpg )
print("ronaldo goals:",  ronaldo_gpg)
```

Since messi has greater goals per game than ronaldo, messi is better.
