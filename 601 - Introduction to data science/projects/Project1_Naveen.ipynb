```python
import pandas as pd
import numpy as np
import csv
from datetime import datetime, date
import warnings
from statistics import mode
```

```python
#********Q1*********

crashesdf = pd.read_csv("crashes.csv",low_memory=False)
personsdf = pd.read_csv('persons.csv',low_memory=False)
vehiclesdf = pd.read_csv('vehicles.csv',low_memory=False)

crashesdf1  = crashesdf[crashesdf['YEAR']==2020]
personsdf1  = personsdf[personsdf['YEAR']==2020]
vehiclesdf1 = vehiclesdf[vehiclesdf['YEAR']==2020]

crashesdfn  = pd.DataFrame(crashesdf1,columns =['REPORT_NO','ACC_DATE','ACC_TIME','WEATHER_DESC','COUNTY_DESC','LATITUDE','LONGITUDE']) 
personsdfn  = pd.DataFrame(personsdf1,columns=['REPORT_NO','SEX_DESC','DATE_OF_BIRTH','LICENSE_STATE_CODE'])
vehiclesdfn = pd.DataFrame(vehiclesdf1,columns=['REPORT_NO','VEH_YEAR','VEH_MAKE','SPEED_LIMIT'])                           

df1 = pd.merge(crashesdfn,personsdfn,on='REPORT_NO') 
df2 = pd.merge(df1,vehiclesdfn,on='REPORT_NO')
df2.drop_duplicates()
df2.dropna(how="any",axis=0,inplace=True)

df2.to_csv('md_crash_2020.csv')
df = pd.read_csv("md_crash_2020.csv",low_memory=False)
```

```python
#**********Q2************
#a
df['DATE_OF_BIRTH']= pd.to_datetime(df['DATE_OF_BIRTH'],format='%Y%m%d')
def age(born):
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
df['Age'] = df['DATE_OF_BIRTH'].apply(age)
```

```python
#**********Q4*******
#a
def haversine_distance(lat1, lon1, lat2, lon2):
   r = 6371
   phi1 = np.radians(lat1)
   phi2 = np.radians(lat2)
   delta_phi = np.radians(lat2-lat1)
   delta_lambda = np.radians(lon2-lon1)
   a = np.sin(delta_phi / 2)**2 + np.cos(phi1) * np.cos(phi2) *   np.sin(delta_lambda / 2)**2
   res = r * (2 * np.arctan2(np.sqrt(a), np.sqrt(1-a)))
   return np.round(res, 2)
df['distance_range'] = haversine_distance(39.016149, -77.117942,df['LATITUDE'],df['LONGITUDE'])
print("Accidents that happened in a circle centered at (39.016149, -77.117942) with a radius of 5 km are: ")
df[df['distance_range']<=5]
```

```python
#b
speed51 = df.apply(lambda x : True
            if x['SPEED_LIMIT'] > 51 else False, axis = 1)
count1 = len(speed51[speed51 == True].index)
print("No of Accidents where the Speed Limit above 51 Km  per hour are:",count1)
#c
speed41 = df.apply(lambda x : True
            if x['SPEED_LIMIT'] < 41 else False, axis = 1)
count2 = len(speed41[speed41 == True].index)
print("No of Accidents where the Speed Limit below 41 Km  per hour are:",count2)
```

```python
#************Q7************
maryland = pd.read_csv("https://raw.githubusercontent.com/simsekergun/DATA601/main/2021Fall/Projects/MD2020_Rain_Snow.csv",low_memory=False)
```

```python
#a
maryland = maryland.rename(columns={'Date':'ACC_DATE'})
maryland['ACC_DATE'] = pd.to_datetime(maryland['ACC_DATE'])
dfn = pd.merge(df,maryland,on='ACC_DATE')
snowy_fun = dfn.apply(lambda x : True
            if x['Precip in.'] > 0.05 or x['Snow in.'] > 1 else False, axis = 1)
count_snowy = len(snowy_fun[snowy_fun == True].index)
count_nosnowy = len(snowy_fun[snowy_fun == False].index) 
print("Ratio of the number of accidents happened in rainy/snowy days to the number of accidents happened in non-rainy/snowy days: \n",count_snowy/count_nosnowy)
```
